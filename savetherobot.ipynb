```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = "https://johnhumigration.aubizconsulting.com.au/"
html = urlopen(url)
```

```python
print(html)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
title = soup.title
print(title)
```

```python
text = soup.get_text()
print(soup.text)
```

```python
soup.find_all('a')
```

```python
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```
